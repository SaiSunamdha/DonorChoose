```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Function to generate and save word cloud from a text file
def generate_wordcloud(file_path, output_image):
    # Read text
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()

    # Generate word cloud
    wordcloud = WordCloud(width=800, height=400, background_color='white', collocations=False).generate(text)

    # Display the word cloud
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(output_image.replace('_', ' ').replace('.png', '').capitalize())
    plt.show()

    # Save to file
    wordcloud.to_file(output_image)

# Generate for approved essays
generate_wordcloud('approved_essays.txt', 'approved_wordcloud.png')

# Generate for rejected essays
generate_wordcloud('rejected_essays.txt', 'rejected_wordcloud.png')
```

```python
# ==== OUTLIER DETECTION ====

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Prepare features of interest
if "total_cost" not in df.columns:
    df["total_cost"] = df["price"] * df["quantity"]

if "num_resources" not in df.columns:
    # Count how many rows (resources) per project_id
    df["num_resources"] = df.groupby("id")["id"].transform("count")

features_to_check = ["total_cost", "num_resources"]

# 2. Function to detect outliers with IQR rule
def detect_outliers(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    outliers = series[(series < lower) | (series > upper)]
    return outliers, lower, upper

for col in features_to_check:
    if col not in df.columns:
        continue
    outliers, low, high = detect_outliers(df[col])
    print(f"\n--- {col} ---")
    print(f"Lower bound: {low:.2f}, Upper bound: {high:.2f}")
    print(f"Outliers detected: {len(outliers)} ({len(outliers)/len(df)*100:.2f}%)")
    
    # Boxplot
    plt.figure(figsize=(8,4))
    sns.boxplot(x=df[col], color="lightblue")
    plt.title(f"Boxplot of {col}")
    plt.show()
    
    # Histogram
    plt.figure(figsize=(8,4))
    sns.histplot(df[col], bins=50, kde=True, color="orange")
    plt.axvline(high, color="red", linestyle="--", label="Upper bound")
    plt.axvline(low, color="red", linestyle="--")
    plt.title(f"Distribution of {col} (with IQR bounds)")
    plt.legend()
    plt.show()

# 3. Handling Options (example: winsorization)
from scipy.stats.mstats import winsorize

df["total_cost_winsor"] = winsorize(df["total_cost"], limits=[0.01, 0.01])  # trim 1% tails
df["num_resources_winsor"] = winsorize(df["num_resources"], limits=[0.01, 0.01])

print("\nCreated winsorized versions: total_cost_winsor, num_resources_winsor")
```

```python
# ==== HYPOTHESIS TESTING: Project Category vs Approval ====

import pandas as pd
from scipy.stats import chi2_contingency

# Ensure approval is int/binary
df["project_is_approved"] = df["project_is_approved"].astype(int)

# Use main category (from 'project_subject_categories')
cat_col = "project_subject_categories"
df[cat_col] = df[cat_col].astype(str).str.split(",").str[0].str.strip()  # take first/main category

# Build contingency table
contingency_table = pd.crosstab(df[cat_col], df["project_is_approved"])
print("Contingency Table (first 10 categories):")
print(contingency_table.head(10))

# Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("\nChi-square Test Results:")
print(f"Chi2 statistic = {chi2:.2f}")
print(f"Degrees of freedom = {dof}")
print(f"P-value = {p:.5f}")

# Interpretation
alpha = 0.05
if p < alpha:
    print(" Reject H: Approval rate is significantly associated with project category.")
else:
    print(" Fail to reject H: No significant association found between approval and category.")
```

```python
# ==== HYPOTHESIS TESTING: Teacher Experience vs Approval ====

import numpy as np
from scipy.stats import ttest_ind

# Ensure approval column is int
df["project_is_approved"] = df["project_is_approved"].astype(int)

# Define groups
first_time = df.loc[df["teacher_number_of_previously_posted_projects"] == 0, "project_is_approved"]
experienced = df.loc[df["teacher_number_of_previously_posted_projects"] > 0, "project_is_approved"]

print(f"First-time teachers: {len(first_time)} projects")
print(f"Experienced teachers: {len(experienced)} projects")

# Compute mean approval rates
print("\nMean Approval Rates:")
print(f"First-time teachers: {first_time.mean():.3f}")
print(f"Experienced teachers: {experienced.mean():.3f}")

# Perform two-sample t-test (unequal variance = Welchs t-test)
t_stat, p_val = ttest_ind(first_time, experienced, equal_var=False)

print("\nT-test Results:")
print(f"T-statistic = {t_stat:.3f}")
print(f"P-value = {p_val:.5f}")

# Interpretation
alpha = 0.05
if p_val < alpha:
    print(" Reject H: Approval rates differ significantly between first-time and experienced teachers.")
else:
    print(" Fail to reject H: No significant difference in approval rates between groups.")
```

```python
# ==== HYPOTHESIS TESTING: Project Cost vs Approval Likelihood ====

from scipy.stats import pearsonr

# Ensure total_cost is present
if "total_cost" not in df.columns:
    df["total_cost"] = df["price"] * df["quantity"]

# Approval as int (0/1)
df["project_is_approved"] = df["project_is_approved"].astype(int)

# Compute Pearson correlation
valid_data = df.dropna(subset=["total_cost", "project_is_approved"])
corr, p_val = pearsonr(valid_data["total_cost"], valid_data["project_is_approved"])

print("Pearson Correlation Results:")
print(f"Correlation coefficient (r) = {corr:.4f}")
print(f"P-value = {p_val:.5f}")

# Interpretation
alpha = 0.05
if p_val < alpha:
    if corr > 0:
        print(" Reject H: Significant positive correlation  higher costs tend to increase approval likelihood.")
    else:
        print(" Reject H: Significant negative correlation  higher costs tend to decrease approval likelihood.")
else:
    print(" Fail to reject H: No significant correlation between project cost and approval likelihood.")
```

```python
# ==== HYPOTHESIS TESTING: Essay Length vs Approval ====

import numpy as np
from scipy.stats import ttest_ind

# 1. Compute essay length (word count)
df["essay_length"] = df["combined_essays"].fillna("").apply(lambda x: len(str(x).split()))

# 2. Split into groups
approved = df.loc[df["project_is_approved"] == 1, "essay_length"]
rejected = df.loc[df["project_is_approved"] == 0, "essay_length"]

print(f"Approved projects: {len(approved)} essays")
print(f"Rejected projects: {len(rejected)} essays")

print("\nMean Essay Lengths:")
print(f"Approved: {approved.mean():.1f} words")
print(f"Rejected: {rejected.mean():.1f} words")

# 3. Perform Welchs t-test (unequal variances)
t_stat, p_val = ttest_ind(approved, rejected, equal_var=False)

print("\nT-test Results:")
print(f"T-statistic = {t_stat:.3f}")
print(f"P-value = {p_val:.5f}")

# 4. Interpretation
alpha = 0.05
if p_val < alpha:
    print(" Reject H: Essay length differs significantly between approved and rejected projects.")
    if approved.mean() > rejected.mean():
        print("    Longer essays are associated with higher approval likelihood.")
    else:
        print("    Shorter essays are associated with higher approval likelihood.")
else:
    print(" Fail to reject H: No significant difference in essay length between groups.")
```

```python
# ==== HYPOTHESIS TESTING: Subject Category Bias in Approvals ====

import pandas as pd
from scipy.stats import chi2_contingency

# Use main subject category (if multiple categories per row, take first one)
df["main_subject_category"] = df["project_subject_categories"].astype(str).str.split(",").str[0].str.strip()

# Build contingency table (category  approval)
contingency = pd.crosstab(df["main_subject_category"], df["project_is_approved"])

print("Contingency Table (first 10 categories):")
print(contingency.head(10))

# Run Chi-square test
chi2, p_val, dof, expected = chi2_contingency(contingency)

print("\nChi-square Test Results:")
print(f"Chi2 statistic = {chi2:.2f}")
print(f"Degrees of freedom = {dof}")
print(f"P-value = {p_val:.5f}")

# Interpretation
alpha = 0.05
if p_val < alpha:
    print(" Reject H: Approval rates differ significantly across subject categories (bias detected).")
else:
    print(" Fail to reject H: No significant evidence of bias in approval across subject categories.")

# --- Identify categories with biggest deviations ---
diff = (contingency - expected) / expected  # standardized residuals
diff_df = pd.DataFrame(diff, columns=contingency.columns, index=contingency.index)

print("\nCategories with strongest positive/negative residuals (indicating bias):")
print(diff_df.abs().sum(axis=1).sort_values(ascending=False).head(10))
```

```python
# ===============================================
# Block 3  Fast & Stable ML Modeling Pipeline
# ===============================================

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.preprocessing import OneHotEncoder, StandardScaler, MaxAbsScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB, ComplementNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    classification_report, confusion_matrix, roc_auc_score, roc_curve
)
import matplotlib.pyplot as plt

# Optional models (if installed)
try:
    from xgboost import XGBClassifier
    HAS_XGB = True
except Exception:
    HAS_XGB = False

try:
    from lightgbm import LGBMClassifier
    HAS_LGBM = True
except Exception:
    HAS_LGBM = False

# -----------------------
# 1) Data Processing
# -----------------------

df = df.copy()

# Target
y = df["project_is_approved"].astype(int)

# Safe imputations
df["combined_essays"] = df.get("combined_essays", pd.Series(index=df.index)).fillna("No essay provided").astype(str)
df["project_resource_summary"] = df.get("project_resource_summary", pd.Series(index=df.index)).fillna("No description")

# Feature engineering
df["essay_length"] = df["combined_essays"].apply(lambda x: len(str(x).split()))
df["total_cost"] = pd.to_numeric(df.get("price", 0), errors="coerce").fillna(0) * \
                   pd.to_numeric(df.get("quantity", 0), errors="coerce").fillna(0)

df["teacher_number_of_previously_posted_projects"] = pd.to_numeric(
    df.get("teacher_number_of_previously_posted_projects", 0), errors="coerce"
).fillna(0)
df["teacher_experience"] = (df["teacher_number_of_previously_posted_projects"] > 0).astype(int)

df["project_submitted_datetime"] = pd.to_datetime(df.get("project_submitted_datetime"), errors="coerce")
df["submit_month"] = df["project_submitted_datetime"].dt.month.fillna(0).astype(int)
df["submit_dow"]   = df["project_submitted_datetime"].dt.dayofweek.fillna(0).astype(int)

if "id" in df.columns:
    df["num_resources"] = df.groupby("id")["id"].transform("count").astype(int)
else:
    df["num_resources"] = 1

# Feature lists
categorical_cols = ["teacher_prefix", "school_state", "project_grade_category", "project_subject_categories"]
numeric_cols = ["essay_length", "total_cost", "teacher_experience",
                "teacher_number_of_previously_posted_projects", "submit_month", "submit_dow", "num_resources"]
text_col = "combined_essays"

# Ensure presence/types
for c in categorical_cols:
    if c not in df.columns: df[c] = "Unknown"
    df[c] = df[c].astype(str).fillna("Unknown")

for c in numeric_cols:
    if c not in df.columns: df[c] = 0
    df[c] = pd.to_numeric(df[c], errors="coerce").fillna(0)

df[text_col] = df[text_col].astype(str).fillna("No essay provided")

# Split
X = df[categorical_cols + numeric_cols + [text_col]].copy()
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)

# -----------------------
# 2) Preprocessors
# -----------------------

# Linear models: sparse TF-IDF + StandardScaler on numerics
preprocessor_std = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_cols),
        ("num", StandardScaler(), numeric_cols),
        ("txt", TfidfVectorizer(max_features=5000, stop_words="english"), text_col),
    ]
)

# Naive Bayes: keep non-negative numerics
preprocessor_nb = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_cols),
        ("num", MaxAbsScaler(), numeric_cols),
        ("txt", TfidfVectorizer(max_features=5000, stop_words="english"), text_col),
    ],
    sparse_threshold=1.0
)

# Helper for dense OHE across sklearn versions (1.7 uses sparse_output)
def dense_ohe():
    try:
        return OneHotEncoder(handle_unknown="ignore", sparse_output=False)  # sklearn >=1.2
    except TypeError:
        return OneHotEncoder(handle_unknown="ignore", sparse=False)        # sklearn <=1.1

# Tree-friendly: TF-IDF -> SVD (200 dims) + dense OHE + passthrough numerics
text_svd_pipe = Pipeline([
    ("tfidf", TfidfVectorizer(max_features=3000, stop_words="english")),
    ("svd", TruncatedSVD(n_components=200, random_state=42))
])
preprocessor_tree = ColumnTransformer(
    transformers=[
        ("cat", dense_ohe(), categorical_cols),
        ("num", "passthrough", numeric_cols),
        ("txt", text_svd_pipe, text_col),
    ],
    remainder="drop",
    sparse_threshold=0.0
)

# -----------------------
# 3) Models
# -----------------------

models = {
    "Logistic Regression": Pipeline([
        ("prep", preprocessor_std),
        ("clf", LogisticRegression(max_iter=1000, class_weight="balanced"))
    ]),
    "MultinomialNB": Pipeline([
        ("prep", preprocessor_nb),
        ("clf", MultinomialNB(alpha=1.0))
    ]),
    "ComplementNB": Pipeline([
        ("prep", preprocessor_nb),
        ("clf", ComplementNB(alpha=1.0))
    ]),

    # Trees/Boosting with tree-friendly preprocessor & lighter params
    "Decision Tree": Pipeline([
        ("prep", preprocessor_tree),
        ("clf", DecisionTreeClassifier(
            max_depth=14, min_samples_leaf=10, random_state=42, class_weight="balanced"
        ))
    ]),
    "Random Forest": Pipeline([
        ("prep", preprocessor_tree),
        ("clf", RandomForestClassifier(
            n_estimators=150,            # reduced from 300
            max_depth=18,                # cap depth
            min_samples_leaf=10,         # regularize
            max_features="sqrt",
            bootstrap=True,
            max_samples=0.6,             # only 60% rows per tree  much faster
            n_jobs=-1,
            random_state=42,
            class_weight="balanced_subsample"
        ))
    ]),
    "Gradient Boosting": Pipeline([
        ("prep", preprocessor_tree),
        ("clf", GradientBoostingClassifier(
            random_state=42, max_depth=3, subsample=0.8
        ))
    ]),
}

if HAS_XGB:
    models["XGBoost"] = Pipeline([
        ("prep", preprocessor_tree),
        ("clf", XGBClassifier(
            n_estimators=300, max_depth=6, learning_rate=0.08,
            subsample=0.8, colsample_bytree=0.8, reg_lambda=1.0,
            objective="binary:logistic", eval_metric="auc",
            n_jobs=-1, random_state=42
        ))
    ])

if HAS_LGBM:
    models["LightGBM"] = Pipeline([
        ("prep", preprocessor_tree),
        ("clf", LGBMClassifier(
            n_estimators=500, num_leaves=63, learning_rate=0.05,
            subsample=0.8, colsample_bytree=0.8, random_state=42, n_jobs=-1
        ))
    ])

# -----------------------
# 4) Evaluation Helper
# -----------------------

def evaluate_model(name, pipeline, X_train, y_train, X_test, y_test, do_cv=True, cv_splits=5):
    print("\n" + "="*30)
    print(f"Model: {name}")
    print("="*30)

    # Reduce CV folds for heavy models
    if do_cv:
        splits = 3 if name in {"Random Forest", "Gradient Boosting", "XGBoost", "LightGBM"} else cv_splits
        cv = StratifiedKFold(n_splits=splits, shuffle=True, random_state=42)
        cv_auc = cross_val_score(pipeline, X_train, y_train, cv=cv, scoring="roc_auc", n_jobs=-1)
        print(f"Cross-validated ROC-AUC: {cv_auc.mean():.3f} (+/- {cv_auc.std():.3f})")

    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_test)

    if hasattr(pipeline[-1], "predict_proba"):
        y_prob = pipeline.predict_proba(X_test)[:, 1]
    else:
        scores = pipeline.decision_function(X_test)
        y_prob = (scores - scores.min()) / (scores.max() - scores.min() + 1e-12)

    acc  = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, zero_division=0)
    rec  = recall_score(y_test, y_pred, zero_division=0)
    f1   = f1_score(y_test, y_pred, zero_division=0)
    auc  = roc_auc_score(y_test, y_prob)

    print("\nMetrics on Test Set")
    print(f"Accuracy : {acc:.3f}")
    print(f"Precision: {prec:.3f}")
    print(f"Recall   : {rec:.3f}")
    print(f"F1-score : {f1:.3f}")
    print(f"ROC-AUC  : {auc:.3f}")

    print("\nClassification Report:\n", classification_report(y_test, y_pred, digits=3))
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

    return y_prob

# -----------------------
# 5) Run & Compare
# -----------------------

plt.figure(figsize=(7,6))
roc_plotted = False

for name, pipe in models.items():
    try:
        y_prob = evaluate_model(name, pipe, X_train, y_train, X_test, y_test, do_cv=True, cv_splits=5)
        fpr, tpr, _ = roc_curve(y_test, y_prob)
        plt.plot(fpr, tpr, label=name)
        roc_plotted = True
    except Exception as e:
        print(f"[{name}] Skipped due to error:\n  {e}")

if roc_plotted:
    plt.plot([0, 1], [0, 1], "k--")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curves  Model Comparison")
    plt.legend()
    plt.tight_layout()
    plt.show()
else:
    print("No ROC curves plotted (all models failed?).")
```

```python
# ===============================================
# Block 3  Fast & Stable ML Modeling Pipeline
# (with HistGradientBoosting for speed)
# ===============================================

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.preprocessing import OneHotEncoder, StandardScaler, MaxAbsScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB, ComplementNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    classification_report, confusion_matrix, roc_auc_score, roc_curve
)
import matplotlib.pyplot as plt

# Optional models (if installed)
try:
    from xgboost import XGBClassifier
    HAS_XGB = True
except Exception:
    HAS_XGB = False

try:
    from lightgbm import LGBMClassifier
    HAS_LGBM = True
except Exception:
    HAS_LGBM = False

# -----------------------
# 1) Data Processing
# -----------------------

df = df.copy()

# Target
y = df["project_is_approved"].astype(int)

# Safe imputations
df["combined_essays"] = df.get("combined_essays", pd.Series(index=df.index)).fillna("No essay provided").astype(str)
df["project_resource_summary"] = df.get("project_resource_summary", pd.Series(index=df.index)).fillna("No description")

# Feature engineering
df["essay_length"] = df["combined_essays"].apply(lambda x: len(str(x).split()))
df["total_cost"] = pd.to_numeric(df.get("price", 0), errors="coerce").fillna(0) * \
                   pd.to_numeric(df.get("quantity", 0), errors="coerce").fillna(0)

df["teacher_number_of_previously_posted_projects"] = pd.to_numeric(
    df.get("teacher_number_of_previously_posted_projects", 0), errors="coerce"
).fillna(0)
df["teacher_experience"] = (df["teacher_number_of_previously_posted_projects"] > 0).astype(int)

df["project_submitted_datetime"] = pd.to_datetime(df.get("project_submitted_datetime"), errors="coerce")
df["submit_month"] = df["project_submitted_datetime"].dt.month.fillna(0).astype(int)
df["submit_dow"]   = df["project_submitted_datetime"].dt.dayofweek.fillna(0).astype(int)

if "id" in df.columns:
    df["num_resources"] = df.groupby("id")["id"].transform("count").astype(int)
else:
    df["num_resources"] = 1

# Feature lists
categorical_cols = ["teacher_prefix", "school_state", "project_grade_category", "project_subject_categories"]
numeric_cols = ["essay_length", "total_cost", "teacher_experience",
                "teacher_number_of_previously_posted_projects", "submit_month", "submit_dow", "num_resources"]
text_col = "combined_essays"

# Ensure presence/types
for c in categorical_cols:
    if c not in df.columns: df[c] = "Unknown"
    df[c] = df[c].astype(str).fillna("Unknown")

for c in numeric_cols:
    if c not in df.columns: df[c] = 0
    df[c] = pd.to_numeric(df[c], errors="coerce").fillna(0)

df[text_col] = df[text_col].astype(str).fillna("No essay provided")

# Split
X = df[categorical_cols + numeric_cols + [text_col]].copy()
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)

# -----------------------
# 2) Preprocessors
# -----------------------

# Linear models: sparse TF-IDF + StandardScaler on numerics
preprocessor_std = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_cols),
        ("num", StandardScaler(), numeric_cols),
        ("txt", TfidfVectorizer(max_features=5000, stop_words="english"), text_col),
    ]
)

# Naive Bayes: keep non-negative numerics
preprocessor_nb = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_cols),
        ("num", MaxAbsScaler(), numeric_cols),
        ("txt", TfidfVectorizer(max_features=5000, stop_words="english"), text_col),
    ],
    sparse_threshold=1.0
)

# Helper for dense OHE across sklearn versions (1.7 uses sparse_output)
def dense_ohe():
    try:
        return OneHotEncoder(handle_unknown="ignore", sparse_output=False)  # sklearn >=1.2
    except TypeError:
        return OneHotEncoder(handle_unknown="ignore", sparse=False)        # sklearn <=1.1

# Tree-friendly: TF-IDF -> SVD (160 dims) + dense OHE + passthrough numerics
text_svd_pipe = Pipeline([
    ("tfidf", TfidfVectorizer(max_features=2500, stop_words="english")),
    ("svd", TruncatedSVD(n_components=160, random_state=42))
])
preprocessor_tree = ColumnTransformer(
    transformers=[
        ("cat", dense_ohe(), categorical_cols),
        ("num", "passthrough", numeric_cols),
        ("txt", text_svd_pipe, text_col),
    ],
    remainder="drop",
    sparse_threshold=0.0
)

# -----------------------
# 3) Models
# -----------------------

models = {
    "Logistic Regression": Pipeline([
        ("prep", preprocessor_std),
        ("clf", LogisticRegression(max_iter=1000, class_weight="balanced"))
    ]),
    "MultinomialNB": Pipeline([
        ("prep", preprocessor_nb),
        ("clf", MultinomialNB(alpha=1.0))
    ]),
    "ComplementNB": Pipeline([
        ("prep", preprocessor_nb),
        ("clf", ComplementNB(alpha=1.0))
    ]),

    # Decision Tree (fast)
    "Decision Tree": Pipeline([
        ("prep", preprocessor_tree),
        ("clf", DecisionTreeClassifier(
            max_depth=14, min_samples_leaf=10, random_state=42, class_weight="balanced"
        ))
    ]),

    # Random Forest (tuned for speed)
    "Random Forest": Pipeline([
        ("prep", preprocessor_tree),
        ("clf", RandomForestClassifier(
            n_estimators=150,
            max_depth=18,
            min_samples_leaf=10,
            max_features="sqrt",
            bootstrap=True,
            max_samples=0.6,
            n_jobs=-1,
            random_state=42,
            class_weight="balanced_subsample"
        ))
    ]),

    # >>> HistGradientBoosting (fast GBDT) with early stopping <<<
    "HistGradientBoosting": Pipeline([
        ("prep", preprocessor_tree),
        ("clf", HistGradientBoostingClassifier(
            learning_rate=0.08,
            max_depth=6,               # shallow trees keep it fast
            max_leaf_nodes=31,         # small trees
            l2_regularization=0.0,
            max_bins=255,
            early_stopping=True,
            validation_fraction=0.1,   # holdout for early stop
            n_iter_no_change=10,       # stop if no improvement
            random_state=42
        ))
    ]),
}

if HAS_XGB:
    models["XGBoost"] = Pipeline([
        ("prep", preprocessor_tree),
        ("clf", XGBClassifier(
            n_estimators=300, max_depth=6, learning_rate=0.08,
            subsample=0.8, colsample_bytree=0.8, reg_lambda=1.0,
            objective="binary:logistic", eval_metric="auc",
            n_jobs=-1, random_state=42
        ))
    ])

if HAS_LGBM:
    models["LightGBM"] = Pipeline([
        ("prep", preprocessor_tree),
        ("clf", LGBMClassifier(
            n_estimators=500, num_leaves=63, learning_rate=0.05,
            subsample=0.8, colsample_bytree=0.8, random_state=42, n_jobs=-1
        ))
    ])

# -----------------------
# 4) Evaluation Helper
# -----------------------

def evaluate_model(name, pipeline, X_train, y_train, X_test, y_test, do_cv=True, cv_splits=5):
    print("\n" + "="*30)
    print(f"Model: {name}")
    print("="*30)

    # Heaviest models: smaller CV; HGB: skip CV (uses internal early stopping)
    if do_cv:
        if name in {"Random Forest", "XGBoost", "LightGBM"}:
            splits = 3
        elif name in {"HistGradientBoosting"}:
            splits = 0  # skip CV; rely on early stopping
        else:
            splits = cv_splits

        if splits >= 2:
            cv = StratifiedKFold(n_splits=splits, shuffle=True, random_state=42)
            cv_auc = cross_val_score(pipeline, X_train, y_train, cv=cv, scoring="roc_auc", n_jobs=-1)
            print(f"Cross-validated ROC-AUC: {cv_auc.mean():.3f} (+/- {cv_auc.std():.3f})")
        else:
            print("Cross-validated ROC-AUC: (skipped for speed)")

    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_test)

    if hasattr(pipeline[-1], "predict_proba"):
        y_prob = pipeline.predict_proba(X_test)[:, 1]
    else:
        # HGB exposes predict_proba; if any model lacks it, fallback:
        if hasattr(pipeline[-1], "decision_function"):
            scores = pipeline.decision_function(X_test)
            y_prob = (scores - scores.min()) / (scores.max() - scores.min() + 1e-12)
        else:
            # As a last resort, cast 0/1 preds to probs (not ideal, but safe)
            y_prob = y_pred.astype(float)

    acc  = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, zero_division=0)
    rec  = recall_score(y_test, y_pred, zero_division=0)
    f1   = f1_score(y_test, y_pred, zero_division=0)
    try:
        auc  = roc_auc_score(y_test, y_prob)
    except Exception:
        # If y_prob has only 0/1 from fallback above, guard AUC
        auc = float("nan")

    print("\nMetrics on Test Set")
    print(f"Accuracy : {acc:.3f}")
    print(f"Precision: {prec:.3f}")
    print(f"Recall   : {rec:.3f}")
    print(f"F1-score : {f1:.3f}")
    print(f"ROC-AUC  : {auc:.3f}")

    print("\nClassification Report:\n", classification_report(y_test, y_pred, digits=3))
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

    return y_prob

# -----------------------
# 5) Run & Compare
# -----------------------

plt.figure(figsize=(7,6))
roc_plotted = False

for name, pipe in models.items():
    try:
        y_prob = evaluate_model(name, pipe, X_train, y_train, X_test, y_test, do_cv=True, cv_splits=5)
        # Only plot if y_prob has variation
        if np.unique(y_prob).size > 1:
            fpr, tpr, _ = roc_curve(y_test, y_prob)
            plt.plot(fpr, tpr, label=name)
            roc_plotted = True
    except Exception as e:
        print(f"[{name}] Skipped due to error:\n  {e}")

if roc_plotted:
    plt.plot([0, 1], [0, 1], "k--")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curves  Model Comparison")
    plt.legend()
    plt.tight_layout()
    plt.show()
else:
    print("No ROC curves plotted (all models failed?).")
```
